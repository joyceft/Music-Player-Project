```python
#for .tar.gz file(406)
def untar_dir(srcname, dstPath):
    tarHandle = tarfile.open(srcname, "r:gz")
    for filename in tarHandle.getnames():
        print(filename)
    tarHandle.extractall(dstPath)
    tarHandle.close()   
```

```python
import os
import os.path
import glob
```

```python
def separate_file(file_list, keyword):
    keyword_file = []
    for filename in file_list:
        if str(filename).find(keyword) > 0:
            keyword_file.append(filename)
    return keyword_file
file_list_1 = GetFileName("E:/projects/Music/untar")
play_list = separate_file(file_list_1, "play")
print(play_list[:10])
```

```python
mergefiledir = "E:/projects/Music/untar/"
play = open("E:/projects/Music/untar/play_combine.log", "wb", encoding="utf-8")

for filename in play_list:
    filepath = mergefiledir + filename
    print(filepath)
    for line in open(filepath):
        play.writelines(line)
    play.write('\n')

play.close()
#mergefiledir = os.getcwd() + ""
```
